```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('C:/Users/rapiduser/Box/EPA STAR 2019 (Community Resistance to Environmental Disasters)/Data/Landucci_2012_AST/Landucci_2012_ReleaseofHazardousSubstancesinFloodEventsDamageModelForAtmosphericStorageTanks.csv')
X_train = df.iloc[:,1:3].to_numpy() #D,H
Y_train = df.iloc[:,3:6].to_numpy() #t,C,PCR
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, Y_train)
r_sq = model.score(X_train, Y_train)
intercept, coefficients = model.intercept_, model.coef_
Y_pred = model.predict(X_train)
```

```python
print('coefficient of determination:', r_sq)
print('intercept:', intercept)
print('coefficients:', coefficients, sep='\n')
#print('predicted response:', Y_pred, sep='\n')
```

```python
import numpy as np
import statsmodels.api as sm
model = sm.OLS(Y_train, X_train)
results = model.fit()
results.summary()
```

```python
print('coefficient of determination:', results.rsquared)
print('adjusted coefficient of determination:', results.rsquared_adj)
print('regression coefficients:', results.params)
```
